### We use groups of 50 cities and pause for 60 seconds every time we reach 50 because Open Weather Map only allows 60 calls per minute on their free teir

```python
city_lats = []
for i, city in enumerate(cities):
    if i < 10:
        url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
        # Reference the enumerated city name and replace blank spaces in the name with + to concatenate
        city_url = url + "&q=" + city.replace(" ","+")
        city_weather = requests.get(city_url).json()
        city_lat = city_weather["coord"]["lat"]
        city_lats.append(city_lat)
print(city_lats)
```

```python
# Empty list to hold weather data
city_data = []
# print to mark progress
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1
# Directly from BCS:
# In the code block, we have initialized the counters at 1 because we want the first iteration 
# of the logging for each recorded response and the set to start at 1.

# Loop through all cities and use enumerate method
# This allows us to use the index number and the name of the city that is referenced by that index
for i, city in enumerate(cities):
    
    # Group cities in sets of 50 for "logging" purposes
    # Check if i is 50 or a multiple of 50
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        # Pause the script for 60 seconds so Open Weather Map doesn't boot us
        time.sleep(60)
    
    # Create endpoint url for each city
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
    # Reference the enumerated city name and replace blank spaces in the name with + to concatenate
    city_url = url + "&q=" + city.replace(" ","+")
    
    # Log the set count, record count, and city
    print(f'Processing record {record_count} of Set {set_count} | {city}')
    # Add 1 to record count
    record_count += 1
    
    # Exception Handling and JSON Data parsing and collection
    try:
        # Make the request
        city_weather = requests.get(city_url).json()
        # Parse out the needed data from the json dictionaries
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        # Convert date to ISO standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime("%Y-%m-%d %H:%M:%S")
        
        # Append city information as dictionary to city_data list
        city_data.append({"City": city.title(), 
                          "Lat":city_lat, 
                          "Lng":city_lng, 
                          "Max Temp":city_max_temp, 
                          "Humidity":city_humidity, 
                          "Cloudiness":city_clouds, 
                          "Wind Speed":city_wind, 
                          "Country":city_country, 
                          "Date":city_date})
    # If error occurs, skip the city, go back to top of loop
    except:
        print("City not found. Skipping ...")
        # General purpose statement to handle all errors
        pass
    
# Indicate that this script and the data loading is complete
print("-------------------------")
print("Data Retrieval Complete")
print("-------------------------")
```
