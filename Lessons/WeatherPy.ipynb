## 6.1.4 New File

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create sets of longitudes and latitudes
lats = np.random.uniform(-90, 90, size=1500)
lngs = np.random.uniform(-180, 180, size=1500)
# Store in memory as pairs
lat_lngs = zip(lats, lngs)
# Return memory object
lat_lngs
```

Above object stores 1500 tuples of latitude-longitude pairs, where each latitude and longitude can be accessed by index 0 and 1, respectively

```python
# Unpack zip object with list function and save to new variable
coordinates = list(lat_lngs)
```

```python
# Loop to display coordinates
for coordinate in coordinates:
    print(coordinate[0], coordinate[1])
```

## 6.1.5

```python
from citipy import citipy
```

```python
# Generate list of cities to around and at least 500
cities = []
# Identify cities by coordinates
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # Conditional to make sure city is not already in list
    if city not in cities:
        cities.append(city)

# Print length of list to check for 500
print(len(cities))
```
