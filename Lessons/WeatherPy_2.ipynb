## 6.3.1 Creating Scatterplots from data

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# Read CSV file created in last lesson
# cities_csv = os.path.join("weather_data/cities.csv")
# Convert CSV to Dataframe
city_data_df = pd.read_csv("weather_data/cities.csv")
```

```python
city_data_df.head(10)
```

```python
# Extract relevant fields from dataframe for plotting
lats = city_data_df["Lat"]
max_temps = city_data_df["Max Temp"]
humidity = city_data_df["Humidity"]
cloudiness = city_data_df["Cloudiness"]
wind_speed = city_data_df["Wind Speed"]
```

### Switch to API_Practice to practice time module

```python
# Scatterplot for latitude vs max temp
import time # to get dates

plt.scatter(lats, max_temps, 
            edgecolor='black', linewidths=1, marker='o', 
            alpha=0.8, label='Cities')

plt.title(f'City Latitude Vs. Max Temperature {time.strftime("%x")}')
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid(True)

# Save figure
plt.savefig("weather_data/Fig1.png")

plt.show()
```
